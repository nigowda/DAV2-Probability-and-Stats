```python
def unstack_col(df, col):
  a = df[col].apply(lambda x:str(x).split(", ")).tolist()
  b = pd.DataFrame(a, index = df['title']).stack().reset_index()
  b.drop("level_1", axis = 1, inplace = True)
  b.rename({0:col}, axis = 1, inplace= True)
  return b
```
